##Calculate Account Age
This block adds the account_age_years column your user_df.

```python
# Convert 'created_at' column to datetime objects
user_df['created_at_dt'] = pd.to_datetime(user_df['created_at'], errors='coerce')

# Standardize the 'created_at' column to UTC
if user_df['created_at_dt'].dt.tz is None:
    user_df['created_at_dt'] = user_df['created_at_dt'].dt.tz_localize('utc')
else:
    user_df['created_at_dt'] = user_df['created_at_dt'].dt.tz_convert('utc')

# Get the current time, explicitly in UTC
now_utc = pd.Timestamp.now(tz="utc")

# Calculate account age in years
user_df['account_age_years'] = (now_utc - user_df['created_at_dt']).dt.days / 365.25

print("--- Account Age (Years) Distribution ---")
print(user_df['account_age_years'].describe())
```

##Definition of "True Novice" & "True Expert" Cohorts

Methodology: Defining "True Novice" & "True Expert" Cohorts
The Approach: Data-Driven Quartile Analysis
The goal of this analysis was to answer the challenge question, "Who adopts Coding Agents on GitHub (e.g., newcomers vs. experienced developers)?" To do this, we first had to define "newcomer" (Novice) and "experienced developer" (Expert).

Since "expertise" is not a column in the dataset, we used two proxies to create a robust definition:

*   *followers*: A proxy for a user's reputation and impact in the community.
*   *account_age_years*: A proxy for a user's time and experience on the GitHub platform.

We used a quartile analysis on the user_df (which contains all 1,796 users in the AIDev-pop dataset) to find the 25th percentile (Q1) and 75th percentile (Q3) for both of these proxy variables.

Running the *.describe()* function on these columns yielded the following thresholds:



*   **Followers:**
    *  25th Percentile (Q1): 31.0
    *  75th Percentile (Q3): 178.0     

* **Account Age (Years):**
    *  25th Percentile (Q1): 7.76
    *  75th Percentile (Q3): 13.68

The Definitions
We defined our two cohorts by selecting the users at the extreme ends of this analysis. This is to ensure we are comparing two distinct groups and get the clearest possible signal.

"True Novice" (Newcomer): A user was assigned to this cohort if they were in the bottom 25% for both variables:

followers â‰¤ 31.0

AND

account_age_years â‰¤ 7.76

"True Expert" (Experienced Developer): A user was assigned to this cohort if they were in the top 25% for both variables:

followers â‰¥ 178.0

AND

account_age_years â‰¥ 13.68

This approach gave us our two final cohorts for comparison: 297 "True Novice" users (who submitted 5,928 AI-assisted PRs) and 182 "True Expert" users (who submitted 1,612 AI-assisted PRs).

```python
# --- Configuration: These are our data-driven thresholds ---
FOLLOWER_NOVICE_THRESHOLD = 31.0
FOLLOWER_EXPERT_THRESHOLD = 178.0
AGE_NOVICE_THRESHOLD = 7.76
AGE_EXPERT_THRESHOLD = 13.68

# Get user_ids for "True Novices"
novice_user_ids = user_df[
    (user_df['followers'] <= FOLLOWER_NOVICE_THRESHOLD) &
    (user_df['account_age_years'] <= AGE_NOVICE_THRESHOLD)
]['id'].unique()

# Get user_ids for "True Experts"
expert_user_ids = user_df[
    (user_df['followers'] >= FOLLOWER_EXPERT_THRESHOLD) &
    (user_df['account_age_years'] >= AGE_EXPERT_THRESHOLD)
]['id'].unique()

# --- Create the final PR DataFrames for each cohort ---
novice_prs = pr_df[pr_df['user_id'].isin(novice_user_ids)]
expert_prs = pr_df[pr_df['user_id'].isin(expert_user_ids)]

print("--- Cohorts Defined ---")
print(f"Found {len(novice_user_ids)} 'True Novice' users, with {len(novice_prs)} PRs.")
print(f"Found {len(expert_user_ids)} 'True Expert' users, with {len(expert_prs)} PRs.")
```

##Task Type Analysis (Finding 1)

```python
# Merge with task types
novice_tasks = pd.merge(novice_prs, pr_task_type_df)
expert_tasks = pd.merge(expert_prs, pr_task_type_df)

print("\n--- 'TRUE NOVICE' Task Distribution (Percentages) ---")
print(novice_tasks['type'].value_counts(normalize=True) * 100)

print("\n--- 'TRUE EXPERT' Task Distribution (Percentages) ---")
print(expert_tasks['type'].value_counts(normalize=True) * 100)
```

##Acceptance vs. Rejection Rate Analysis (Finding 2)
This block answers "What is the quality of their PRs?"

```python
print("--- Analysis of PR Outcomes (Acceptance vs. Rejection Rates) ---")

# --- Novice Outcome Analysis ---
novice_closed_prs = novice_prs[novice_prs['state'] == 'closed']
total_novice_closed_count = len(novice_closed_prs)
merged_novice_count = novice_closed_prs['merged_at'].count()
rejected_novice_count = total_novice_closed_count - merged_novice_count
novice_acceptance_rate = (merged_novice_count / total_novice_closed_count) * 100
novice_rejection_rate = (rejected_novice_count / total_novice_closed_count) * 100

print(f"\n--- 'TRUE NOVICE' PR Outcomes ---")
print(f"Total Closed PRs: {total_novice_closed_count}")
print(f"Acceptance Rate: {novice_acceptance_rate:.2f}%")
print(f"Rejection Rate: {novice_rejection_rate:.2f}%")

# --- Expert Outcome Analysis ---
expert_closed_prs = expert_prs[expert_prs['state'] == 'closed']
total_expert_closed_count = len(expert_closed_prs)
merged_expert_count = expert_closed_prs['merged_at'].count()
rejected_expert_count = total_expert_closed_count - merged_expert_count
expert_acceptance_rate = (merged_expert_count / total_expert_closed_count) * 100
expert_rejection_rate = (rejected_expert_count / total_expert_closed_count) * 100

print(f"\n--- 'TRUE EXPERT' PR Outcomes ---")
print(f"Total Closed PRs: {total_expert_closed_count}")
print(f"Acceptance Rate: {expert_acceptance_rate:.2f}%")
print(f"Rejection Rate: {expert_rejection_rate:.2f}%")
```

##PR Size Analysis (Finding 3)
This block answers "How large are their PRs?"

```python
print("--- PR Size Analysis ---")

# Aggregate additions/deletions from the commit details
pr_size_stats = pr_commit_details_df.groupby('pr_id').agg(
    total_additions=('additions', 'sum'),
    total_deletions=('deletions', 'sum')
).reset_index()
pr_size_stats['total_changes'] = pr_size_stats['total_additions'] + pr_size_stats['total_deletions']

# Merge size data with our cohorts
novice_prs_with_size = pd.merge(novice_prs, pr_size_stats, left_on='id', right_on='pr_id')
expert_prs_with_size = pd.merge(expert_prs, pr_size_stats, left_on='id', right_on='pr_id')

print("\n--- 'TRUE NOVICE' PR Size (Statistics) ---")
print(novice_prs_with_size['total_changes'].describe())

print("\n--- 'TRUE EXPERT' PR Size (Statistics) ---")
print(expert_prs_with_size['total_changes'].describe())
```

##Commit Granularity Analysis (Finding 4)

```python
print("--- Commit Granularity Analysis ---")

# Count commits for each PR
pr_commit_counts = pr_commits_df.groupby('pr_id').size().reset_index(name='commit_count')

# Merge commit counts with our cohorts
novice_prs_with_commits = pd.merge(novice_prs, pr_commit_counts, left_on='id', right_on='pr_id')
expert_prs_with_commits = pd.merge(expert_prs, pr_commit_counts, left_on='id', right_on='pr_id')

print("\n--- 'TRUE NOVICE' Commit Granularity (Statistics) ---")
print(novice_prs_with_commits['commit_count'].describe())

print("\n--- 'TRUE EXPERT' Commit Granularity (Statistics) ---")
print(expert_prs_with_commits['commit_count'].describe())
```

##Agent Preference by Task (Finding 7)
this block answers "Which Agent is prefered per task type"

```python
print("--- Agent Preference by Task Type Analysis ---")

# Find the correct 'agent' column name (e.g., 'agent_x' or 'agent')
AGENT_COLUMN_NAME = 'agent_x' if 'agent_x' in novice_tasks.columns else 'agent'

# --- 'TRUE NOVICE' Analysis ---
novice_agent_by_task = (novice_tasks.groupby('type')[AGENT_COLUMN_NAME]
                        .value_counts(normalize=True)
                        .mul(100)
                        .rename('percentage')
                        .reset_index())
novice_agent_by_task.rename(columns={AGENT_COLUMN_NAME: 'agent'}, inplace=True)

# --- 'TRUE EXPERT' Analysis ---
expert_agent_by_task = (expert_tasks.groupby('type')[AGENT_COLUMN_NAME]
                        .value_counts(normalize=True)
                        .mul(100)
                        .rename('percentage')
                        .reset_index())
expert_agent_by_task.rename(columns={AGENT_COLUMN_NAME: 'agent'}, inplace=True)

# --- Print The Full Results ---
pd.set_option('display.max_rows', 100)

print("\n" + "="*40)
print("   'TRUE NOVICE' Agent Use by Task (All Tasks)")
print("="*40)
print(novice_agent_by_task)

print("\n" + "="*40)
print("   'TRUE EXPERT' Agent Use by Task (All Tasks)")
print("="*40)
print(expert_agent_by_task)
```

```python
# --- 1. Prepare Data for Heatmaps ---
# We use the 'novice_tasks' and 'expert_tasks' DataFrames
# And we must find the correct agent column name (e.g., 'agent_x')
AGENT_COLUMN_NAME = 'agent_x' if 'agent_x' in novice_tasks.columns else 'agent'

# Novice heatmap data
novice_heatmap_data = (novice_tasks.groupby('type')[AGENT_COLUMN_NAME]
                        .value_counts(normalize=True)
                        .mul(100)
                        .rename('percentage')
                        .reset_index())
novice_heatmap_pivot = novice_heatmap_data.pivot(
    index='type', columns=AGENT_COLUMN_NAME, values='percentage'
).fillna(0) # Fill tasks they didn't do with 0

# Expert heatmap data
expert_heatmap_data = (expert_tasks.groupby('type')[AGENT_COLUMN_NAME]
                        .value_counts(normalize=True)
                        .mul(100)
                        .rename('percentage')
                        .reset_index())
expert_heatmap_pivot = expert_heatmap_data.pivot(
    index='type', columns=AGENT_COLUMN_NAME, values='percentage'
).fillna(0) # Fill tasks they didn't do with 0


# --- 2. Create the Visualization (2 subplots) ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 8), gridspec_kw={'width_ratios': [1, 1]})
fig.suptitle('Finding 4: Novices and Experts Specialize with Different Agents', fontsize=16, y=1.03)

# --- Plot 1: Novice Heatmap ---
sns.heatmap(novice_heatmap_pivot,
            annot=True,     # Show percentages in cells
            fmt=".1f",      # Format to 1 decimal place
            cmap="Reds",    # Color map
            linewidths=.5,  # Add lines between cells
            ax=ax1)
ax1.set_title('True Novice: Agent Use by Task (%)')
ax1.set_ylabel('Task Type')
ax1.set_xlabel('Agent')

# --- Plot 2: Expert Heatmap ---
sns.heatmap(expert_heatmap_pivot,
            annot=True,
            fmt=".1f",
            cmap="Blues",   # Use a different color
            linewidths=.5,
            ax=ax2)
ax2.set_title('True Expert: Agent Use by Task (%)')
ax2.set_ylabel('') # Hide y-label, it's redundant
ax2.set_xlabel('Agent')

plt.tight_layout()
plt.show()
```
