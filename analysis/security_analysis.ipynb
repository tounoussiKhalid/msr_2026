```python
import pandas as pd

# Define the base path for loading from Hugging Face
base_path = "hf://datasets/hao-li/AIDev/"

try:
    # --- 1. Load Agentic-PR Data (from AIDEV-POP) ---
    # Load main agent PR table
    agent_pr_df = pd.read_parquet(f"{base_path}pull_request.parquet")
    # Load task types for agent PRs
    agent_task_df = pd.read_parquet(f"{base_path}pr_task_type.parquet")

    # --- 2. Load Human-PR Data (The comparison set) ---
    # Load main human PR table
    human_pr_df = pd.read_parquet(f"{base_path}human_pull_request.parquet")
    # Load task types for human PRs
    human_task_df = pd.read_parquet(f"{base_path}human_pr_task_type.parquet")

    print(f"Loaded {len(agent_pr_df)} Agentic-PRs for comparison.")
    print(f"Loaded {len(human_pr_df)} Human-PRs for comparison.\n")

    # --- Analysis 1: Task Type Distribution (Replicates data for Table 4) ---
    print("--- Analysis 1: Task Type Distribution ---")

    # Merge agent PRs with their tasks
    # The agent PR 'id' links to the task 'pr_id'
    agent_tasks_merged = pd.merge(agent_pr_df, agent_task_df, left_on='id', right_on='pr_id')

    # Merge human PRs with their tasks
    # The human PR 'id' links to the task 'pr_id'
    human_tasks_merged = pd.merge(human_pr_df, human_task_df, left_on='id', right_on='pr_id')

    # Get normalized value counts for each
    agent_task_dist = agent_tasks_merged['type'].value_counts(normalize=True).rename("Agent (%)") * 100
    human_task_dist = human_tasks_merged['type'].value_counts(normalize=True).rename("Human (%)") * 100

    # Combine into one DataFrame for easy comparison
    task_comparison_df = pd.concat([human_task_dist, agent_task_dist], axis=1).fillna(0)
    print(task_comparison_df.sort_values(by="Human (%)", ascending=False))
    print("\n")


    # --- Analysis 2: Overall Acceptance Rate (Replicates data for Table 5) ---
    print("--- Analysis 2: Overall Acceptance Rate ---")

    # Acceptance rate is the percentage of PRs where 'merged_at' is not null
    agent_acceptance_rate = (agent_pr_df['merged_at'].notna().sum() / len(agent_pr_df)) * 100
    human_acceptance_rate = (human_pr_df['merged_at'].notna().sum() / len(human_pr_df)) * 100

    print(f"Human-PR Acceptance Rate: {human_acceptance_rate:.2f}%")
    print(f"Agentic-PR Acceptance Rate: {agent_acceptance_rate:.2f}%")

    # This confirms Finding #2: Agents lag human acceptance rates
    # The preprint reports 76.8% for Human vs. 65.3% (Codex) down to 38.2% (Copilot)
    # Our overall agent rate will be an average of these.

except Exception as e:
    print(f"Error loading data: {e}")
    print("Please ensure you have network access and the 'pyarrow' and 'fsspec' libraries installed.")
```
