```python
import pandas as pd
import os
import subprocess
import json
import shutil  # For cleaning up directories
import time

# --- Configuration ---

# 1. Define the base path for loading from Hugging Face
BASE_PATH = "hf://datasets/hao-li/AIDev/"

# 2. Define a local path to clone repositories into
# Use an absolute path to avoid all relative path errors
CLONE_DIR = os.path.abspath("./cloned_repos")
os.makedirs(CLONE_DIR, exist_ok=True)
print(f"Repositories will be cloned into: {CLONE_DIR}")


# 3. Define your SAST tool command (Example for Semgrep)
# We want JSON output to a specific file
SAST_COMMAND = [
    "semgrep",
    "scan",
    "--config", "auto",  # Use auto-detected rules
    "--json",
    "-o", "scan_results.json",
    "." # Scan the current directory
]
```

```python
import pandas as pd
import os
import subprocess
import json
import shutil
import time

# (This block now contains the fixes)

def clone_repo(repo_full_name):
    """
    Clones a repository given its full name (e.g., 'user/repo').
    Returns the local path to the cloned repo.
    """
    repo_url = f"https://github.com/{repo_full_name}.git"
    local_repo_path = os.path.join(CLONE_DIR, repo_full_name.replace("/", "_"))

    if os.path.exists(local_repo_path):
        return local_repo_path

    try:
        print(f"Cloning {repo_full_name}...")
        subprocess.run(
            ["git", "clone", "--depth", "1", repo_url, local_repo_path],
            check=True, capture_output=True, text=True, timeout=300
        )
        return local_repo_path
    except subprocess.CalledProcessError as e:
        print(f"Failed to clone {repo_full_name}: {e.stderr}")
        return None
    except subprocess.TimeoutExpired:
        print(f"Clone timed out for {repo_full_name}")
        return None

def run_sast_scan(repo_path, files_to_scan):
    """
    Runs the pre-configured SAST command in the repo_path
    on *only* the specified files.
    """
    scan_output_file = os.path.join(repo_path, "scan_results.json")

    if os.path.exists(scan_output_file):
        os.remove(scan_output_file)

    # --- FIX: Build a dynamic command to scan only specific files ---
    # Start with the base command
    dynamic_sast_command = [
        "semgrep", "scan", "--config", "auto",
        "--json", "-o", "scan_results.json"
    ]

    # Add only the files that exist
    existing_files_to_scan = []
    for f in files_to_scan:
        if os.path.exists(os.path.join(repo_path, f)):
            existing_files_to_scan.append(f)
        else:
            print(f"  File not found, skipping: {f}")

    if not existing_files_to_scan:
        print("  No changed files exist in this checkout. Skipping scan.")
        return set()

    dynamic_sast_command.extend(existing_files_to_scan)
    # --- End Fix ---

    try:
        subprocess.run(
            dynamic_sast_command, # Use the dynamic command
            cwd=repo_path,
            check=True, capture_output=True, text=True, timeout=300
        )

        if os.path.exists(scan_output_file):
            with open(scan_output_file, 'r') as f:
                results = json.load(f)
            os.remove(scan_output_file)
            return set(finding['check_id'] for finding in results.get('results', []))
        else:
            return set()

    except subprocess.CalledProcessError as e:
        if os.path.exists(scan_output_file):
             with open(scan_output_file, 'r') as f:
                results = json.load(f)
             os.remove(scan_output_file)
             return set(finding['check_id'] for finding in results.get('results', []))

        print(f"  SAST scan failed: {e.stderr}")
        return set()
    except subprocess.TimeoutExpired:
        print(f"  SAST scan timed out (even on few files).")
        return set()

def get_vulnerability_diff(local_repo_path, commit_sha):
    """
    Checks out the 'before' and 'after' state of a commit,
    scans *only changed files*, and returns the vulnerability diff.
    """
    original_cwd = os.getcwd()
    os.chdir(local_repo_path)

    introduced_vulns, mitigated_vulns = set(), set()

    try:
        print(f"  Fetching commit {commit_sha[:7]}...")
        subprocess.run(
            ["git", "fetch", "origin", commit_sha, "--depth=2"],
            check=True, capture_output=True, text=True, timeout=300
        )

        before_vulns = set()
        changed_files = []

        try:
            # --- This block runs for NON-ROOT commits ---
            # Get the list of changed files by comparing commit to its parent
            diff_cmd = ["git", "diff-tree", "--no-commit-id", "--name-only", "-r", f"{commit_sha}^", commit_sha]
            diff_result = subprocess.run(diff_cmd, check=True, capture_output=True, text=True)
            changed_files = [f for f in diff_result.stdout.strip().split('\n') if f]

            if not changed_files:
                print(f"  No files changed in commit {commit_sha[:7]}. Skipping.")
                return set(), set()

            # "Before" Scan
            print(f"  Checking out parent of {commit_sha[:7]} for 'before' scan...")
            subprocess.run(["git", "checkout", f"{commit_sha}^"], check=True, capture_output=True, text=True)
            before_vulns = run_sast_scan(local_repo_path, changed_files)

        except subprocess.CalledProcessError as e:
            # --- This block runs for ROOT commits (which have no parent '^') ---
            if "did not match" in e.stderr:
                print(f"  Commit {commit_sha[:7]} is a root commit. 'Before' state is empty.")
                before_vulns = set() # Before state is empty

                # Get the list of files *created* in this root commit
                diff_cmd = ["git", "diff-tree", "--no-commit-id", "--name-only", "-r", commit_sha]
                diff_result = subprocess.run(diff_cmd, check=True, capture_output=True, text=True)
                changed_files = [f for f in diff_result.stdout.strip().split('\n') if f]
            else:
                print(f"  Error checking out parent: {e.stderr}")
                raise e # Re-raise other unexpected errors

        if not changed_files:
            print(f"  No files to scan for commit {commit_sha[:7]}. Skipping.")
            return set(), set()

        # --- "After" Scan ---
        print(f"  Checking out commit {commit_sha[:7]} for 'after' scan...")
        subprocess.run(["git", "checkout", commit_sha], check=True, capture_output=True, text=True)
        after_vulns = run_sast_scan(local_repo_path, changed_files)

        introduced_vulns = after_vulns - before_vulns
        mitigated_vulns = before_vulns - after_vulns

    except subprocess.CalledProcessError as e:
        print(f"  Git operation failed for {commit_sha}: {e.stderr}")
    except subprocess.TimeoutExpired:
        print(f"  Git fetch/checkout timed out for {commit_sha}")
    finally:
        subprocess.run(["git", "checkout", "main"], capture_output=True, text=True)
        os.chdir(original_cwd)

    return introduced_vulns, mitigated_vulns

print("Helper functions defined successfully (with timeout fix).")
```

```python
print("Creating analysis cohorts...")

# 1. Merge task types into PR tables
agent_prs_with_tasks = pd.merge(agent_pr_df, agent_task_df)
human_prs_with_tasks = pd.merge(human_pr_df, human_task_df)
print(human_prs_with_tasks.columns)
print(human_prs_with_tasks.columns)

# 2. Filter for 'fix' PRs
agent_fix_prs = agent_prs_with_tasks[agent_prs_with_tasks['type'] == 'fix']
human_fix_prs = human_prs_with_tasks[human_prs_with_tasks['type'] == 'fix']

print(f"Agent 'fix' PRs: {len(agent_fix_prs)}")
print(f"Human 'fix' PRs: {len(human_fix_prs)}")

# 3. Sample the data for a manageable run (e.g., 20 PRs from each)
# NOTE: This may result in more than 20 commits, as a PR can have multiple commits
agent_sample_prs = agent_fix_prs.sample(n=20, random_state=42)
human_sample_prs = human_fix_prs.sample(n=20, random_state=42)

# 4. Add a 'source' column for aggregation later
agent_sample_prs['source'] = 'Agent'
human_sample_prs['source'] = 'Human'

combined_sample_prs = pd.concat([agent_sample_prs, human_sample_prs])

# 5. Merge to get repo_full_name
sample_with_repos = pd.merge(combined_sample_prs, repo_df[['id', 'full_name']], left_on='repo_id', right_on='id')

print( sample_with_repos.columns)
print( pr_commits_df.columns)

# 6. Merge to get commit 'sha'
# This is our final table to iterate over
final_sample_commits = pd.merge(sample_with_repos, pr_commits_df[['pr_id', 'sha']], left_on='id_x', right_on='pr_id')

print(final_sample_commits.columns)

print(f"--- Ready for Analysis ---")
print(f"Total PRs in sample: {len(combined_sample_prs)}")
print(f"Total Commits to scan: {len(final_sample_commits)}")
print(final_sample_commits[['id_x','sha', 'source', 'full_name']].head())
```

```python
# --- Corrected Block 5A: Fetch Human Commit SHAs ---

print("--- Starting Block 5A: Fetching Human PR Commits via API ---")

# We have 'agent_sample_prs' and 'human_sample_prs' from Block 5

# 1. Get the AGENT commits (from the pre-loaded pr_commits_df)
# We need to get the full_name from the repo_url first
agent_sample_prs['full_name'] = agent_sample_prs['repo_url'].apply(lambda x: '/'.join(x.split('/')[-2:]))
print( agent_sample_prs.columns)

# 'id_x' is the correct PR ID after the merge in Block 5
agent_commits_df = pd.merge(
    agent_sample_prs,
    pr_commits_df[['pr_id', 'sha']],
    left_on='id',
    right_on='pr_id'
)
agent_commits_df = agent_commits_df[['pr_id', 'sha', 'source', 'full_name']]


# 2. Fetch HUMAN commits using the GitHub API
human_commits_list = []
for _, row in human_sample_prs.iterrows():
    # --- THIS IS THE FIX ---
    # Use repo_url (which exists) instead of repo_id (which doesn't)
    repo_url = row['repo_url']
    pr_number = row['number']

    # Extract repo_full_name from 'https://api.github.com/repos/user/repo'
    repo_full_name = '/'.join(repo_url.split('/')[-2:])
    # --- END FIX ---

    print(f"Fetching commits for Human PR: {repo_full_name}#{pr_number}")

    # Use 'gh' (GitHub CLI) to get PR commits as JSON
    api_command = [
        "gh", "pr", "view", str(pr_number),
        "--repo", repo_full_name,
        "--json", "commits"
    ]

    try:
        result = subprocess.run(api_command, check=True, capture_output=True, text=True, timeout=60)
        pr_data = json.loads(result.stdout)

        for commit in pr_data.get('commits', []):
            human_commits_list.append({
                "pr_id": row['id'], # The PR id for humans is 'id'
                "sha": commit['oid'],
                "source": "Human",
                "full_name": repo_full_name
            })

    except subprocess.CalledProcessError as e:
        print(f"  Failed to fetch commits for {repo_full_name}#{pr_number}: {e.stderr}")
    except Exception as e:
        print(f"  An error occurred: {e}")

human_commits_df = pd.DataFrame(human_commits_list)

# 3. Combine both dataframes
if not human_commits_df.empty:
    final_sample_commits = pd.concat([agent_commits_df, human_commits_df]).reset_index(drop=True)
    print("\n--- Cohorts Successfully Rebuilt ---")
    print(f"Total Agent commits in sample: {len(agent_commits_df)}")
    print(f"Total Human commits in sample: {len(human_commits_df)}")
    print(f"Total commits to scan: {len(final_sample_commits)}")
    print("\nSample of final data:")
    print(final_sample_commits.sample(n=5))
else:
    print("\n--- FAILED TO FETCH HUMAN COMMITS ---")
    print("Please check API rate limits or gh installation.")
    print("Stopping before Block 6.")
    # Set this to prevent Block 6 from running on bad data
    final_sample_commits = pd.DataFrame()
```

```python
# --- New Block 5A: Fetch Human Commit SHAs ---

print("--- Starting Block 5A: Fetching Human PR Commits via API ---")

# We already have 'human_sample_prs' and 'agent_sample_prs' from Block 5
# But we need the 'sha' for the human PRs.


# 1. Get the agent commits (like before)
agent_commits_df = pd.merge(agent_sample_prs, pr_commits_df[['pr_id', 'sha']], left_on='id', right_on='pr_id')
agent_commits_df = pd.merge(agent_commits_df, repo_df[['id', 'full_name']], left_on='repo_id', right_on='id')
agent_commits_df = agent_commits_df[['pr_id', 'sha', 'source', 'full_name']]

print(human_sample_prs.columns)

# 2. Fetch Human commits using the GitHub API
human_commits_list = []
for _, row in human_sample_prs.iterrows():
    # We need the repo full_name and the PR number
    repo_id = row['repo_id']
    pr_number = row['number']

    # Get the repo's full_name
    repo_full_name = repo_df[repo_df['id'] == repo_id]['full_name'].values[0]

    print(f"Fetching commits for Human PR: {repo_full_name}#{pr_number}")

    # Use 'gh' (GitHub CLI) to get PR commits as JSON
    # This is a robust way to use the API
    api_command = [
        "gh", "pr", "view", str(pr_number),
        "--repo", repo_full_name,
        "--json", "commits"
    ]

    try:
        result = subprocess.run(api_command, check=True, capture_output=True, text=True, timeout=60)
        pr_data = json.loads(result.stdout)

        for commit in pr_data.get('commits', []):
            human_commits_list.append({
                "pr_id": row['id'],
                "sha": commit['oid'],
                "source": "Human",
                "full_name": repo_full_name
            })

    except subprocess.CalledProcessError as e:
        print(f"  Failed to fetch commits for {repo_full_name}#{pr_number}: {e.stderr}")
    except Exception as e:
        print(f"  An error occurred: {e}")

human_commits_df = pd.DataFrame(human_commits_list)

# 3. Combine both dataframes
if not human_commits_df.empty:
    final_sample_commits = pd.concat([agent_commits_df, human_commits_df]).reset_index(drop=True)
    print("\n--- Cohorts Successfully Rebuilt ---")
    print(f"Total Agent commits in sample: {len(agent_commits_df)}")
    print(f"Total Human commits in sample: {len(human_commits_df)}")
    print(f"Total commits to scan: {len(final_sample_commits)}")
    print("\nSample of final data:")
    print(final_sample_commits.sample(n=5))
else:
    print("\n--- FAILED TO FETCH HUMAN COMMITS ---")
    print("Please check API rate limits or gh installation.")
    print("Stopping before Block 6.")
    # Set this to prevent Block 6 from running on bad data
    final_sample_commits = pd.DataFrame()
```

```python
print(f"--- Starting Analysis on {len(final_sample_commits)} total commits ---")

results = []
start_time = time.time()

# 7. Run the core "BYOD" loop
for _, row in final_sample_commits.iterrows():
    print(f"\nAnalyzing PR {row['pr_id']} (Commit {row['sha'][:7]}...) from {row['full_name']}")

    # Step A: Clone the repo
    local_repo_path = clone_repo(row['full_name'])
    if not local_repo_path:
        print(f"  Skipping commit, repo clone failed.")
        continue # Skip if clone failed

    # Step B: Get the vulnerability diff
    introduced, mitigated = get_vulnerability_diff(local_repo_path, row['sha'])

    # Step C: Store results
    results.append({
        "pr_id": row['pr_id'],
        "sha": row['sha'],
        "source": row['source'], # 'Agent' or 'Human'
        "num_introduced": len(introduced),
        "num_mitigated": len(mitigated),
        "introduced_types": list(introduced),
        "mitigated_types": list(mitigated)
    })
    print(f"  Done. Introduced: {len(introduced)}, Mitigated: {len(mitigated)}")

end_time = time.time()
print(f"\n--- Analysis Complete in {end_time - start_time:.2f} seconds ---")
```

```python
if not results:
    print("No results were generated. Please check for errors in the loop.")
else:
    results_df = pd.DataFrame(results)

    print("\n--- Raw Results (First 5) ---")
    print(results_df.head())

    print("\n--- Summary: Vulnerabilities per Commit ---")
    # Show per-commit summary
    print(results_df.groupby('source')[['num_introduced', 'num_mitigated']].mean())

    print("\n--- Summary: Vulnerabilities per PR ---")
    # Show per-PR summary (a PR might have multiple commits)
    pr_summary = results_df.groupby(['source', 'pr_id'])[['num_introduced', 'num_mitigated']].sum()
    print(pr_summary.groupby('source').mean())
```
